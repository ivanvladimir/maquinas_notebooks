```python
from maquinas.regular.dfa import DeterministicFiniteAutomaton as DFA
from maquinas.simulation import Simulation
```

```python
tres_bes=DFA(Q=['q0','q1','q2','q3','q4'],
                         sigma=['a','b'],
                         q_0='q0',
                         A=['q3','q4'],
                         delta=[
                            (('q0','a'),'q4'),
                            (('q0','b'),'q1'),
                            (('q1','a'),'q1'),
                            (('q1','b'),'q2'),
                            (('q2','a'),'q2'),
                            (('q2','b'),'q3'),
                            (('q3','a'),'q3'),
                            (('q3','b'),'q1'),
                            (('q4','a'),'q4'),
                            (('q4','b'),'q1')
                         ])
```
