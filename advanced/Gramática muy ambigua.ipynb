```python
from maquinas.contextfree.cfg import ContextFreeGrammar as CFG
from maquinas.parser.earley_parser import print_chart, chart2table
```

```python
epsilon_heavy=CFG("S->SS; S->a; S-> epsilon")
```

```python
print(epsilon_heavy.summary())
```

```python
s="aa"
roots,chart,forest=epsilon_heavy.parse(s)
print(f'Con "{s}" Se acepta?', "Sí" if len(roots)>0 else "No" )
```

```python
epsilon_heavy.print_chart(s,chart)
```

```python
epsilon_heavy.graph_forest(forest,span=True,show_id=True)
```

```python
trees=list(epsilon_heavy.extract_trees(forest,max_ancesters=3))
epsilon_heavy.graph_trees(trees,show_id=True)
```
