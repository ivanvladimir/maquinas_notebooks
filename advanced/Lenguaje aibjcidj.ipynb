```python
from maquinas.recursivelyenumerable.tspda import TwoStackPushDownAutomaton
from maquinas.simulation import Simulation
```

```python
# Ejercicio uno
ww=TwoStackPushDownAutomaton(Q=['q_0','q_1','q_2','q_3','q_4'],
         sigma=['a','b'],
         gamma=['A','B'],
         q_0='q_0',
         A=['q_4'],
         delta=[
            (('q_0','a','Z0','Z0'),[('q_0','AZ0','Z0')]),
            (('q_0','a','A','Z0'),[('q_0','AA','Z0')]),
            (('q_0','c','A','Z0'),[('q_2','epsilon','Z0')]),
            (('q_0','epsilon','Z0','Z0'),[('q_4','Z0','Z0')]),
            
            (('q_0','b','A','Z0'),[('q_1','A','BZ0')]),
            (('q_1','b','A','B'),[('q_1','A','BB')]),
            (('q_0','b','Z0','Z0'),[('q_1','Z0','BZ0')]),
            (('q_1','b','Z0','B'),[('q_1','Z0','BB')]),
            (('q_1','d','Z0','B'),[('q_3','Z0','epsilon')]),
             
            (('q_1','c','A','B'),[('q_2','epsilon','B')]),
            (('q_2','c','A','B'),[('q_2','epsilon','B')]),
            (('q_2','c','A','Z0'),[('q_2','epsilon','Z0')]),
            (('q_2','epsilon','Z0','Z0'),[('q_3','Z0','Z0')]),
            
            (('q_2','d','Z0','B'),[('q_3','Z0','epsilon')]),
            (('q_3','d','Z0','B'),[('q_3','Z0','epsilon')]), 
            
            (('q_3','epsilon','Z0','Z0'),[('q_4','Z0','Z0')]),
        ]
    )
```

```python
for q,a,w_ in ww.delta_stepwise(""):
    if a:
        print(f"{a} -> {ww.states2string(q)}", end=",\n ")
    else:
        print(f"{ww.states2string(q)}",end="\n ")
    res=q
    
print(f"\nCon {ww.states2string(res)} Se acepta?", "Sí" if ww.acceptor(res) else "No" )
```
