```python
i3_lessthan_j=CFG('R-> SaT ;S->aSaaa; S-> B; T-> aT; T-> epsilon; B-> bB; B-> epsilon')
print(i3_lessthan_j.summary())
```

```python
roots,chart,forest=i3_lessthan_j.parse('aabaaaaaaa')
trees=i3_lessthan_j.extract_trees(forest)
for i,step in enumerate(i3_lessthan_j.derivation(list(trees)[0])):
    if not i:
        print("".join(step),end="")
    else:
        print(" ⇒ ","".join(step),end="\n ")
```
