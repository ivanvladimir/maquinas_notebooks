```python
import os
import sys
from maquinas.regular.dfa import DeterministicFiniteAutomaton as DFA
from maquinas.simulation import Simulation
```

```python
# Creating automata finito
ceros_uno=DFA(Q=['q_0','q_1'],
                         sigma=['0','1'],
                         q_0='q_0',
                         A=['q_1'],
                         delta=[
                            (('q_0','0'),'q_0'),
                            (('q_0','1'),'q_1')
                         ])
ceros_uno.graph()
```

```python
ceros_uno.replace_symbol("0","‚öΩ")
ceros_uno.replace_symbol("1","ü•Ö")
ceros_uno.graph()
```

```python
ceros_uno.replace_state("q_0","q‚ÇÄ")
ceros_uno.replace_state("q_1","q‚ÇÅ")
ceros_uno.graph()
```

```python
ceros_uno.add_error_state("q‚Çë")
ceros_uno.graph()
```
