Ejmplifica el uso de emojis

```python
import os
import sys
from maquinas.regular.dfa import DeterministicFiniteAutomaton as DFA
from maquinas.simulation import Simulation
```

```python
# Creating automata finito
ceros_uno=DFA(Q=['q_0','q_1'],
                         sigma=['0','1'],
                         q_0='q_0',
                         A=['q_1'],
                         delta=[
                            (('q_0','0'),'q_0'),
                            (('q_0','1'),'q_1')
                         ])
ceros_uno.graph()
```

```python
ceros_uno.replace_symbol("0","⚽")
ceros_uno.replace_symbol("1","🥅")
ceros_uno.graph()
```

```python
ceros_uno.replace_state("q_0","q₀")
ceros_uno.replace_state("q_1","q₁")
ceros_uno.graph()
```

```python
ceros_uno.add_error_state("qₑ")
ceros_uno.graph()
```

```python
# Creating automata finito
uno_ceros=DFA(Q=['q_0','q_1'],
                         sigma=['0','1'],
                         q_0='q_0',
                         A=['q_1'],
                         delta=[
                            (('q_0','1'),'q_1'),
                            (('q_1','0'),'q_1')
                         ])
uno_ceros.graph()
```

```python
uno_ceros.replace_symbol("0","🐘")
uno_ceros.replace_symbol("1","🐪")
uno_ceros.replace_state("q_0","q₀")
uno_ceros.replace_state("q_1","q₁")
uno_ceros.add_error_state("qₑ")
uno_ceros.graph()
```
